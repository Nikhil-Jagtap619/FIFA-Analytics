## FIFA Data on players (2018)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fifa = pd.read_csv("fifa_data.csv").fillna(0)
fifa.head(7)
```

### Histogram

```python
#histogram on overall skills
bins = np.arange(40,110,10)
plt.style.use("seaborn")
plt.figure(figsize=(8,5),dpi=120)
plt.hist(fifa["Overall"], bins=bins, color="#bdb2ff")
plt.xticks(bins)
plt.title("Skill Set Distribution in FIFA")
plt.xlabel("Overall Rating")
plt.ylabel("Number of players")
plt.show()
```

```python
high_level_player = fifa.loc[fifa["Overall"]>=90]
high_level_player.head(7)
```

### Pie Chart

#### What foot is preferred by the players

```python

left_foot = len(fifa.loc[fifa["Preferred Foot"]=="Left"])
right_foot = len(fifa.loc[fifa["Preferred Foot"]=="Right"])
plt.style.use("seaborn")
plt.figure(figsize=(8,8), dpi=90)
plt.pie([left_foot,right_foot], labels=["left_F","Right_F"], autopct="%.1f%%")
plt.title("Foot Preference")
plt.legend()
plt.show()
```

#### Weight of Players 

```python
fifa.Weight = [int(x.strip("lbs")) if type(x)== str else x for x in fifa.Weight]
under_weight = fifa.loc[fifa["Weight"]<140].count()[0]
medium_weight = fifa.loc[(fifa["Weight"]>=140) & (fifa["Weight"]<170)].count()[0]
heavy_weight = fifa.loc[(fifa["Weight"]>=170) & (fifa["Weight"]<200)].count()[0]
over_weight = fifa.loc[fifa["Weight"]>=200].count()[0]

weights = [under_weight, medium_weight, heavy_weight, over_weight]
labels = ["below 140","140-170","171-200","above 200"]
plt.style.use("seaborn")
plt.figure(figsize=(8,8), dpi=90)
plt.pie(weights, labels=labels, autopct="%.1f%%")
plt.title("Weight distribution of FIFA players (in lbs) ")
plt.legend()
plt.show()
##plt.style.available change your style
```

### Box & Whiskers Chart


#### We willl compare clubs and their highest rating players

```python
barcelona = fifa.loc[fifa["Club"]=="FC Barcelona"]["Overall"]
juventus = fifa.loc[fifa["Club"]=="Juventus"]["Overall"]
mc_united = fifa.loc[fifa["Club"]=="Manchester United"]["Overall"]

plt.figure(figsize=(5,8), dpi=90)
labels = ["Barcelona","Juventus","MC United"]
colors = ["#427aa1","#ced4da","#f28482"]
boxes = plt.boxplot([barcelona,juventus,mc_united], labels=labels, patch_artist=True,medianprops = {"linewidth":2, "color": "#fff3b0"})
for box in boxes["boxes"]:
    box.set(color="#6c757d", linewidth=1.2) #set color of egdes
    box.set(facecolor = colors.pop(0))
    
plt.ylabel("Overall Rating")
plt.title("Comparision of clubs")
plt.show()
```
